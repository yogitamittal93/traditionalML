```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
students=pd.read_csv('/content/student-mat.csv')
```

```python
df = pd.read_csv('/content/student-mat.csv', sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
cat_cols = df.select_dtypes(include='object').columns
```

```python
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
df_encoded.info()
```

```python
x=df_encoded.drop(['G3'], axis=1)
```

```python
df_encoded['G3'] = (df_encoded['G3'] >= 10).astype(int)
```

```python
y=df_encoded['G3']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
model=tf.keras.Sequential([
    tf.keras.Input(shape=(41,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='sigmoid'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
y_train
```

```python
print(x_train.dtypes.value_counts())
```

```python
print(y_train.dtype)
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
# Convert boolean columns in x_train and x_test to integers (0 or 1)
for col in x_train.select_dtypes(include='bool').columns:
    x_train[col] = x_train[col].astype(int)
for col in x_test.select_dtypes(include='bool').columns:
    x_test[col] = x_test[col].astype(int)

# Ensure all columns are numeric, coercing any remaining non-numeric (though there shouldn't be any now)
x_train = x_train.apply(pd.to_numeric, errors='coerce')
x_test = x_test.apply(pd.to_numeric, errors='coerce')

print("x_train dtypes after conversion:")
display(x_train.dtypes)
print("x_test dtypes after conversion:")
display(x_test.dtypes)
```

```python
model.evaluate(x_test, y_test)
```

```python
data = pd.read_csv('/content/student-mat.csv', sep=';')
```

```python
x= data.drop(['G1', 'G2', 'G3'], axis=1)
```

```python
y=data['G3']
```

```python
y=pd.cut(y, bins=[0, 5, 10, 15, 18, 20], labels=['D', 'C', 'B', 'A', 'A+'], right=True)
```

```python
y
```

```python
y=pd.get_dummies(y)
```

```python
x=pd.get_dummies(x, columns=cat_cols, drop_first=True)
```

```python
x.info()
```

```python
for col in x.select_dtypes(include='bool').columns:
    x[col] = x[col].astype(int)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model=tf.keras.Sequential([
    tf.keras.Input(shape=(39,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
model.evaluate(x_test, y_test)
```
